```python
will_drop = ['Est Dia in M(min)','Est Dia in M(max)','Est Dia in Miles(min)','Est Dia in Miles(max)','Est Dia in Feet(min)','Est Dia in Feet(max)',
            'Relative Velocity km per hr','Miles per hour','Miss Dist.(lunar)','Miss Dist.(miles)','Miss Dist.(Astronomical)']
```

```python
df.drop(labels=will_drop,axis=1,inplace=True)
```

```python
drops = ['Neo Reference ID','Equinox','Orbiting Body','Close Approach Date','Name']
```

```python
df.drop(labels=drops,axis=1,inplace=True)
```

```python
df.drop(['Orbit Determination Date'],axis=1,inplace=True)
```

```python
df.columns = df.columns.str.replace(' ','_')
df.head(2)
```

```python
from sklearn.preprocessing import LabelBinarizer
coder = LabelBinarizer()
```

```python
df["Hazardous"] = coder.fit_transform(df["Hazardous"])
```

```python
df.head()
```

```python
plt.figure(figsize=(8, 12))
heatmap = sns.heatmap(df.corr()[['Hazardous']].sort_values(by='Hazardous', ascending=False), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Features Correlating with Hazardous', fontdict={'fontsize':18}, pad=16);
```

```python
X = df.drop("Hazardous",axis=1)
y = df.Hazardous
```

```python
from sklearn.preprocessing import RobustScaler 
```

```python
scaler = RobustScaler()
X_columns = X.columns
X = scaler.fit_transform(X)
X = pd.DataFrame(X,columns=X_columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,stratify=y,test_size=0.2,random_state=123)
```

```python
from xgboost import XGBClassifier 
```

```python
xgb = XGBClassifier(use_label_encoder=False)
xgb_pred_train = xgb.fit(X_train,y_train).predict(X_train)
xgb_pred_test = xgb.predict(X_test)
```

```python
xgb_prob_pred = xgb.predict_proba(X_test)
```

```python
xgb_prob_pred
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(xgb_pred_train,y_train))
```

```python
print(classification_report(xgb_pred_test,y_test))
```

```python
from skopt import BayesSearchCV
from skopt.space import Real, Categorical, Integer 
```

```python
model = XGBClassifier(random_state=123)
```

```python
search_spaces = {'learning_rate':Real(0.01,1.0,'uniform'),
                 'max_depth':Integer(2,12),
                 'subsample':Real(0.1,1.0,'uniform'),
                 'colsample_bytree':Real(0.1,1.0,'uniform'),
                 'reg_lambda':Real(1e-9,100.,'uniform'),
                 'reg_alpha':Real(1e-9,100.,'uniform'),
                 'n_estimators':Integer(50,5000)
    }
```

```python
opt = BayesSearchCV(estimator=model,
                    search_spaces=search_spaces,
                    scoring='accuracy',
                    n_iter=11,
                    n_points=1,
                    n_jobs=1,
                    return_train_score=False,
                    refit=False,
                    optimizer_kwargs={'base_estimator':'GP'},
                    random_state=0)
```

```python
_ =opt.fit(X_train,y_train)
```

```python
print(opt.best_params_)
```

```python
boosted_xgb = XGBClassifier(colsample_bytree=0.432,random_state=123,learning_rate=0.445,max_depth=5,n_estimators=4001,reg_alpha=7.01223,reg_lambda=74.3009,subsample=0.7607)
```

```python
y_prediction_train_boosted = boosted_xgb.fit(X_train,y_train).predict(X_train)
```

```python
print(classification_report(y_prediction_train_boosted,y_train))
```
